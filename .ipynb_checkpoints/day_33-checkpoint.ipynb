```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(2021)
img=np.random.rand(6,6).round(1)
img
```

```python
filter=np.array([[1,0,1],[0,-4,0],[1,0,1]])
```

### 연습문제
- img 에 filter로 convolution 적용한 후,2x2 맥스풀링을 적용하시오.
- convolution 결과에 relu 활성화 함수를 적용하면 어떻게 변할까

```python
result=[]

for i in range(4):
    l=[]
    for j in range(4):
        l.append((img[i:i+3,j:j+3]*filter).sum())
    result.append(l)
    
np.array(result).round(2)
```

```python
result2=np.array(result).copy()
result2[result2<0]=0
result2
```

```python
img = ff2[:,:,0] # R channel 분리
plt.imshow(img, cmap='gray', vmin=0, vmax=1)
```

```python
filter=np.array([[1,0,1],[0,1,0],[1,0,1]])/5
filter
```

```python
w, h = img.shape
conv = []

for i in range(h-3+1):
    l = []
    
    for j in range(w-3+1):
        r = (img[i:i+3, j:j+3]*filter).sum()
        l.append(r)
        
    conv.append(l)
    
conv = np.array(conv, dtype=float)
```
