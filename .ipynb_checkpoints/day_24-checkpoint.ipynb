```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### MNIST 주성분 분석

```python
from sklearn.datasets import load_digits
```

```python
mnist=load_digits()
mnist.keys()
```

```python
mnist.data.shape,mnist.images.shape
```

```python
plt.imshow(mnist.data[100].reshape(8,8),cmap='gray_r',vmin=0,vmax=15)
plt.title(mnist.target[100],fontsize=30)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
pca.fit(mnist.data)
```

```python
pca.components_.shape
```

```python
pca1=pca.components_[0]
pca1.round(2)
```

```python
pca2=pca.components_[1]
pca2.round(2)
```

```python
plt.plot(pca1)
```

```python
plt.imshow(pca1.reshape(8,8))
```

```python
plt.subplot(1,2,1)
plt.imshow(pca1.reshape(8,8))
plt.colorbar()
plt.subplot(1,2,2)
plt.imshow(pca2.reshape(8,8))
plt.colorbar()
```

```python
X_pca=pca.transform(mnist.data)
```

```python
X_pca.shape
```

```python
plt.scatter(X_pca[:,0],X_pca[:,1],alpha=0.2)
```
