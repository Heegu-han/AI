### 연습문제
- 이미지 파일을 불러오면 대부분 0~255 사이의 정수값을 가진다. 이값을 0 ~ 1.0까지의 실수값으로 바꾸시오.
- 10번째 행까지의 값을 모두 1로 바꾸시오.
- 10번째 열까지의 값을 모두 1로 바꾸시오.
- 빨간색과 파란색의 밝기를 절반으로 저정하시오.
- 빨간색 채널의 값을 히스토그램으로 그리시오.
- 흑백이미지로 변환하시오. (채널에 대한 평균값 구하기)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bird=plt.imread('bird.jpg')
print(bird.shape)
plt.imshow(bird)
```

```python
bird[0].shape
```

```python
bird[:,0].shape
```

```python
bird[0,0]
```

```python
bird[:,:,0].shape
```

```python
plt.imshow(bird[:,:,0],cmap='gray',vmin=0,vmax=255)
```

```python
bird[:,:,[0,2]].shape
```

```python
bird.min(),bird.max()
```

```python
bird=bird/255
plt.hist(bird.ravel())
```

```python
bird2=bird.copy()
bird2[:,:100]=1
plt.imshow(bird2)
```

```python
bird2 = bird.copy()
bird2[:, :, [0,2]] = bird2[:, :, [0,2]]/2
plt.imshow(bird2)
```

```python
plt.hist(bird[:,:,0].ravel(), bins=100)
```

```python
bird2=bird.copy()
bird2[50:100,50:400]*=0.5
bird2[bird2>1]=1
plt.imshow(bird2)
```

```python
bird2=bird.copy()
bird2[100:200, 100:200] = bird2[200:300, 200:300]
plt.imshow(bird2)
```

```python
birds=np.array([bird,bird2])
birds.shape
```

```python
plt.imshow(birds[1])
```
